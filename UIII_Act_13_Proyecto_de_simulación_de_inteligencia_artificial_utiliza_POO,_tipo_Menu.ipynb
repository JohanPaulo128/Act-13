#### numero de lista = 04 (Tarjeta de credito)
### Las 5 funciones son:
* realizar_compra
* pagar_deuda
* imprimir_resumen
* asistencia_ia
* puntos_recompensa
### Los 5 atributos son:
* Titular
* Numero
* Limite
* saldo_deuda
* puntos_recompensa

```python
# ==========================================
# DEFINICIÓN DE LA CLASE (EL MOLDE)
# ==========================================
class TarjetaCredito:
    # 1. Función especial para crear los 5 ATRIBUTOS
    def __init__(self, titular, numero, limite):
        self.titular = titular            # Atributo 1
        self.numero = numero              # Atributo 2
        self.limite = limite              # Atributo 3
        self.saldo_deuda = 0.0            # Atributo 4
        self.puntos_recompensa = 0        # Atributo 5

    # --- AQUÍ EMPIEZAN LAS 5 FUNCIONES (MÉTODOS) ---

    # Función 1: Realizar una compra
    def realizar_compra(self, cantidad):
        if cantidad <= (self.limite - self.saldo_deuda):
            self.saldo_deuda += cantidad
            self.puntos_recompensa += 10
            print(f"✅ Compra de ${cantidad} exitosa.")
        else:
            print("❌ Saldo insuficiente en la línea de crédito.")

    # Función 2: Pagar la deuda
    def pagar_deuda(self, abono):
        self.saldo_deuda -= abono
        if self.saldo_deuda < 0: self.saldo_deuda = 0
        print(f"✅ Pago de ${abono} registrado.")

    # Función 3: Mostrar estado en TABLA (Bien alineado)
    def imprimir_resumen(self):
        print("\n" + "="*75)
        # Encabezados con formato de tabla
        print(f"{'TITULAR':<15} | {'NÚMERO':<18} | {'LÍMITE':<10} | {'DEUDA':<10} | {'PUNTOS':<6}")
        print("-" * 75)
        # Datos alineados
        print(f"{self.titular:<15} | {self.numero:<18} | ${self.limite:<9} | ${self.saldo_deuda:<9} | {self.puntos_recompensa:<6}")
        print("="*75 + "\n")

    # Función 4: Asistente de IA (Simula inteligencia de análisis)
    def asistente_ia(self):
        print("🤖 [IA BANCARIA ANALIZANDO...]")
        porcentaje = (self.saldo_deuda / self.limite) * 100
        if porcentaje > 80:
            print("⚠️ Sugerencia: Estás por llegar al límite. ¡No gastes más!")
        elif self.saldo_deuda == 0:
            print("🌟 Sugerencia: Tu historial está impecable. ¡Sigue así!")
        else:
            print("👍 Sugerencia: Tu uso de crédito es moderado y saludable.")

    # Función 5: Canjear puntos
    def canjear_beneficio(self):
        if self.puntos_recompensa >= 50:
            self.puntos_recompensa -= 50
            print("🎁 ¡Felicidades! Canjeaste 50 puntos por un cupón de café.")
        else:
            print(f"⚠️ Te faltan {50 - self.puntos_recompensa} puntos para un premio.")

# ==========================================
# INICIO DEL PROGRAMA (SISTEMA INTERACTIVO)
# ==========================================

print("--- REGISTRO DE NUEVA TARJETA ---")
nombre = input("Nombre del titular: ")
num_t = input("Número de tarjeta (16 dígitos): ")
limite_inicial = float(input("Límite de crédito otorgado: "))

# Creamos el objeto basado en la clase
mi_tarjeta = TarjetaCredito(nombre, num_t, limite_inicial)

# Menú interactivo
opcion = "0"
while opcion != "6":
    print(f"\n--- MENÚ IA - TARJETA DE {nombre.upper()} ---")
    print("1. Ver información (Tabla)")
    print("2. Comprar algo")
    print("3. Pagar deuda")
    print("4. Consultar Asistente IA")
    print("5. Canjear Puntos")
    print("6. Salir")

    opcion = input("Seleccione una opción: ")

    if opcion == "1":
        mi_tarjeta.imprimir_resumen()
    elif opcion == "2":
        monto = float(input("¿Cuánto vas a gastar?: "))
        mi_tarjeta.realizar_compra(monto)
    elif opcion == "3":
        pago = float(input("¿Cuánto vas a pagar a tu tarjeta?: "))
        mi_tarjeta.pagar_deuda(pago)
    elif opcion == "4":
        mi_tarjeta.asistente_ia()
    elif opcion == "5":
        mi_tarjeta.canjear_beneficio()
    elif opcion == "6":
        print("¡Gracias por usar el sistema! Adiós.")
    else:
        print("Opción inválida.")
```
